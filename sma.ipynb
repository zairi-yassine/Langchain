```python
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from dotenv.ipython import load_dotenv
```

```python
load_dotenv(override=True) #force le remplacement de la apikey utilise ici , si une autre existe deja dans notre environement 
```

On va creer un simple agent, on spicifie le model puis on specifie le system prompt, qui depand de la tache que l'agent doit effectuer

```python
agent = create_agent(
    model="openai:gpt-4o",
    system_prompt="You are a helpfull assistant"
)
```

```python
resp = agent.invoke(input={"messages":[
    {"role": "user", "content": "comment je m'appel?"}
]})   #envoie le message a llm
```
